```python
import matplotlib.pyplot as plt

values = [20,30,45,10]
labels =['nokia','samsung','apple','lumia']
colors =['yellow','green','red', 'blue']

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.pie(values,labels=labels, colors=colors,explode=[0.3,0,0,0], shadow=True, startangle=180, autopct='%1.1f%%')
plt.show()
```
